### Creating geometry for police sectors from police beat locations

```python

# Convert beats to geo dataframe
beats_gdf = gpd.GeoDataFrame(beats_df, geometry=gpd.GeoSeries.from_wkt(beats_df['the_geom']))

# Group by 'SECTOR' and dissolve polygons
combined_gdf = beats_gdf.dissolve(by=['DISTRICT', 'SECTOR'])

```

```python
#Dropping old geometry column
combined_gdf = combined_gdf.drop(columns=['the_geom'])

combined_gdf.head(5)
```

```python
total_rows = combined_gdf.shape[0]
print(f"Total number of rows in the combined_gdf: {total_rows}")

#Same number of the sectors in police sentiment data, suggesting this worked appropriately
```

### Creating map of each police beat

```python
from shapely import wkt
import geopandas as gpd

# Create a GeoDataFrame
beats_df['the_geom'] = beats_df['the_geom'].apply(wkt.loads)  # Convert 'the_geom' column to geometries
gdf = gpd.GeoDataFrame(beats_df, geometry='the_geom')
```

```python

# Plotting the data
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, color='lightblue', edgecolor='black')

# Add labels
for idx, row in gdf.iterrows():
    # Extract the centroid of the polygon
    centroid = row['the_geom'].centroid
    # Annotate with the BEAT_NUM
    ax.annotate(text=str(row['BEAT_NUM']), xy=(centroid.x, centroid.y),
                horizontalalignment='center', fontsize=5, color='black')

# Customize the plot
ax.set_title('Beat Numbers')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
plt.show()
```

### Quick police sector cleaning

```python
#Converting Beats into sectors
combined_gdf['SECTOR'] = (combined_gdf['BEAT_NUM'] // 10) * 10
```

```python
combined_gdf
```

```python
#Creating new dataframe that is just police sector and coordinates
police_sector_geom = combined_gdf
police_sector_geom.set_index('SECTOR', inplace=True)
police_sector_geom = combined_gdf.drop(columns=['BEAT', 'BEAT_NUM'])
```

```python
police_sector_geom
```

### Creating a map of each police sector

```python
#Convert to geodataframe
police_sector_geom = gpd.GeoDataFrame(police_sector_geom, geometry='geometry')
```

```python

# Plotting the data
fig, ax = plt.subplots(figsize=(10, 10))
police_sector_geom.plot(ax=ax, color='lightblue', edgecolor='black')

# Add labels
for idx, row in police_sector_geom.iterrows():
    # Extract the centroid of the polygon
    centroid = row.geometry.centroid
    # Annotate with the SECTOR number (which is the index in this case)
    ax.annotate(text=str(idx), xy=(centroid.x, centroid.y),
                horizontalalignment='center', verticalalignment='center', fontsize=5, color='black')

# Customize the plot
ax.set_title('Police Sectors')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
plt.show()
```

### Creating overlay of both maps to ensure location accuracy with convsersion

```python
import matplotlib.pyplot as plt

# Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(12, 12))

# Plot the sectors
police_sector_geom.plot(ax=ax, color='lightblue', edgecolor='black')

# Plot the beats on top
gdf.plot(ax=ax, color='lightgreen', edgecolor='black', alpha=0.2)  # 'alpha' for transparency

# Add labels for beats
for idx, row in gdf.iterrows():
    # Extract the centroid of the polygon
    centroid = row['the_geom'].centroid
    # Annotate with the BEAT_NUM
    ax.annotate(text=str(row['BEAT_NUM']), xy=(centroid.x, centroid.y),
                horizontalalignment='center', fontsize=5, color='black')


# Customize the plot
ax.set_title('Beats and Sectors Overlay')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show the plot
plt.show()
```

## Preparing Police Sentiment Data for Joining

```python
#Renaming and dropping columns. Changing order of each 

# Rename 'START_YEAR_MONTH' to 'Month'
police_df.rename(columns={'START_YEAR_MONTH': 'Month'}, inplace=True)
police_df.rename(columns={'SECTOR': 'Sector'}, inplace=True)
# Drop 'AREA' and 'DISTRICT' columns from police_df
police_df.drop(['AREA', 'DISTRICT'], axis=1, inplace=True)
```

```python
# Sort the DataFrame first by 'Month' and then by 'SECTOR'
police_df = police_df.sort_values(by=['Month', 'Sector'])

# Reset the index of the sorted DataFrame
police_df.reset_index(drop=True, inplace=True)

#Rearrange columns so Month is next to index
# Get a list of all columns except 'Month'
columns = [col for col in police_df.columns if col != 'Month']

# Place 'Month' at the beginning of the column list
new_column_order = ['Month'] + columns

# Reorder the columns in the DataFrame
police_df= police_df[new_column_order]

police_df.head(5)
```

## SQL Code for Joining

```python
#Making new table for Total crime
"""
CREATE TABLE ML_ALLCRIME_EDU_INCOME AS
SELECT *
FROM MACHINE_LEARNING_TOTAL_CRIME
JOIN POLICE_SENTIMENT_JOINING_EDU_INCOME 
USING (Month, Sector);

"""
```

```python
#Making new table for violent crime
"""
CREATE TABLE ML_VIOLENTCRIME AS
SELECT *
FROM MACHINE_LEARNING_VIOLENT_CRIME
JOIN POLICE_SENTIMENT_JOINING_EDU_INCOME 
USING (Month, Sector);

"""
```
