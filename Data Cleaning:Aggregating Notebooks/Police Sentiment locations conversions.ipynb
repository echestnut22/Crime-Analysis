```python
#Removing end_date and converting start_date to just year and month

# Convert START_DATE to datetime
police_df['START_DATE'] = pd.to_datetime(police_df['START_DATE'])

# Extract year and month and then combine as a new column
police_df['START_YEAR_MONTH'] = police_df['START_DATE'].dt.to_period('M')

# Drop the original START_DATE and END_DATE columns
police_df = police_df.drop(columns=['START_DATE', 'END_DATE'])

```

```python
police_df.head(5)
```

```python
year_month_counts = police_df['START_YEAR_MONTH'].value_counts().sort_index()
with pd.option_context('display.max_rows', None):
    print(year_month_counts)
```
