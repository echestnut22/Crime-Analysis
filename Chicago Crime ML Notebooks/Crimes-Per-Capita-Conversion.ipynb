## Joining Census Dataframes

```python
df_census_geo = df_census_geo.rename(columns={'BLOCKCE10': 'CENSUS BLOCK', 'GEOID10': 'CENSUS BLOCK FULL'})
df_census = df_census_pop.merge(df_census_geo[['CENSUS BLOCK FULL', 'the_geom']], on='CENSUS BLOCK FULL', how='left')
df_census
```

```python
import geopandas as gpd
import pandas as pd

# Assuming df_census_pop is your DataFrame with the 'geometry' column
# Clean and validate the 'geometry' column
df_census['the_geom'] = df_census['the_geom'].apply(lambda x: shapely.wkt.loads(x.strip()))  # Strip leading/trailing whitespace

# Create a GeoDataFrame
gdf_census = gpd.GeoDataFrame(df_census, geometry='the_geom')

```

```python
# Spatial join to align geometries and aggregate population data
pop_df = gpd.sjoin(df_sector_geom, gdf_census, how='left', op='intersects')

# Group by 'SECTOR' and aggregate the population data
pop_df = pop_df.groupby('SECTOR')['TOTAL POPULATION'].sum().reset_index()

# Add the 'geometry' column from df_sector_geom
pop_df = pop_df.merge(df_sector_geom[['SECTOR', 'geometry']], on='SECTOR', how='left')
```

```python
pop_df = pop_df.rename(columns={'SECTOR': 'Sector'})

pop_df.head(70)
```

```python
pop_df
```

```python

pop_df = gpd.GeoDataFrame(pop_df, geometry='geometry')

import geopandas as gpd


# Set the initial CRS for the DataFrame (e.g., EPSG:4326 for WGS 84)
initial_crs = 'EPSG:4326'

# Set the CRS for the DataFrame
pop_df.crs = initial_crs

# Define the CRS you want to use for reprojection (e.g., EPSG:3857 for Web Mercator)
desired_crs = 'EPSG:3857'

# Reproject the DataFrame to the desired CRS
pop_df = pop_df.to_crs(desired_crs)


# Calculate area (in square meters) for each sector
pop_df['Area_sqm'] = pop_df['geometry'].area

# Convert 'Area_sqm' to a more suitable unit (e.g., square kilometers)
pop_df['Area_sqkm'] = pop_df['Area_sqm'] / 1e6  # 1 square kilometer = 1,000,000 square meters

# Calculate population density (people per square kilometer)
pop_df['Population Density (people/sqkm)'] = pop_df['TOTAL POPULATION'] / pop_df['Area_sqkm']
```

```python

# Merge violent_df with population_df based on a common identifier, e.g., 'SECTOR' or 'CENSUS BLOCK FULL'
merged_df = df_violent.merge(pop_df, on='Sector', how='left')

# Calculate Violent Crimes per Capita per 1,000 people
merged_df['Violent Crimes per Capita'] = (merged_df['Violent Crimes'] / merged_df['TOTAL POPULATION']) * 1000
```

```python
# Assuming your DataFrame is named pop_df

# List of columns to drop
columns_to_drop = ['geometry', 'Area_sqm', 'Area_sqkm']

# Drop the specified columns
merged_df = merged_df.drop(columns=columns_to_drop)
```
