### Creating GeoDataframe for joining with main dataset

```python
from shapely.geometry import Point

# Function to create a Point object from latitude and longitude
def create_point(lat, lon):
    return Point(lon, lat)

# Function to convert location string to a Point object
def convert_to_point(location_str):
    if pd.isna(location_str):
        return None
    lat, lon = map(float, location_str.strip('()').split(', '))
    return Point(lon, lat)

# Use 'Location' if available, otherwise use 'Latitude' and 'Longitude'
housing_clean['geometry'] = housing_clean.apply(
    lambda row: convert_to_point(row['Location']) if pd.notna(row['Location']) else create_point(row['Latitude'], row['Longitude']),
    axis=1
)

# Create a GeoDataFrame
gdf = gpd.GeoDataFrame(housing_clean, geometry='geometry')





#Dropping columns that don't need for joining table 
columns_to_drop = [
                   'Latitude',
                   'Longitude',
                   'Location']

gdf_housing = gdf.drop(columns=columns_to_drop)

gdf_housing
```

### Joining Datasets. Adding total Units per Sector in new column for analysis

```python
merged_df = gpd.GeoDataFrame(merged_df, geometry='geometry')


# Check and set CRS for 'gdf_housing'
if gdf_housing.crs is None:
    gdf_housing.set_crs("EPSG:4326", inplace=True)

# Check and set CRS for 'merged_df'
if merged_df.crs is None:
    merged_df.set_crs("EPSG:4326", inplace=True)

# Ensure both GeoDataFrames are in the same CRS
gdf_housing = gdf_housing.to_crs(merged_df.crs)

# Perform the spatial join
joined_gdf = gpd.sjoin(gdf_housing, merged_df, how="inner", op="within")

# Group by Sector and aggregate Units
aggregated_data = joined_gdf.groupby('Sector')['Units'].sum().reset_index()

# Merge the aggregated data back with the polygon GeoDataFrame using a left join
merged_df = merged_df.merge(aggregated_data, on='Sector', how='left')

# Replace NaN values in 'Units' with 0
merged_df['Units'].fillna(0, inplace=True)

```

### Cleaning and uploading to database

```python
#Adding standardized numbers for units based on size of sector and population of sector

# Calculate Units per Square Kilometer
merged_df['Units per SqKm'] = merged_df['Units'] / merged_df['Area_sqkm']

# Calculate Units per Capita
merged_df['Units per Capita'] = merged_df['Units'] / merged_df['TOTAL POPULATION']

merged_df.head(5)
```
