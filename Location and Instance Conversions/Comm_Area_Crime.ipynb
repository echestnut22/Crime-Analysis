## Joining Census Dataframes

```python
df_census_geo = df_census_geo.rename(columns={'BLOCKCE10': 'CENSUS BLOCK', 'GEOID10': 'CENSUS BLOCK FULL'})
df_census = df_census_pop.merge(df_census_geo[['CENSUS BLOCK FULL', 'the_geom']], on='CENSUS BLOCK FULL', how='left')
df_census
```

```python
import geopandas as gpd
import pandas as pd

# Assuming df_census_pop is your DataFrame with the 'geometry' column
# Clean and validate the 'geometry' column
df_census['the_geom'] = df_census['the_geom'].apply(lambda x: shapely.wkt.loads(x.strip()))  # Strip leading/trailing whitespace

# Create a GeoDataFrame
gdf_census = gpd.GeoDataFrame(df_census, geometry='the_geom')
```

```python
# Spatial join to align geometries and aggregate population data
pop_df = gpd.sjoin(df_community_area_geom, gdf_census, how='left', op='intersects')

# Group by 'SECTOR' and aggregate the population data
pop_df = pop_df.groupby(['COMMUNITY','AREA_NUM_1'])['TOTAL POPULATION'].sum().reset_index()

# Add the 'geometry' column from df_community_area_geom
pop_df = pop_df.merge(df_community_area_geom[['COMMUNITY','AREA_NUM_1','geometry']], on=['COMMUNITY','AREA_NUM_1'], how='left')
```

```python
pop_df = gpd.GeoDataFrame(pop_df, geometry='geometry')

import geopandas as gpd


# Set the initial CRS for the DataFrame (e.g., EPSG:4326 for WGS 84)
initial_crs = 'EPSG:4326'

# Set the CRS for the DataFrame
pop_df.crs = initial_crs

# Define the CRS you want to use for reprojection (e.g., EPSG:3857 for Web Mercator)
desired_crs = 'EPSG:3857'

# Reproject the DataFrame to the desired CRS
pop_df = pop_df.to_crs(desired_crs)


# Calculate area (in square meters) for each sector
pop_df['Area_sqm'] = pop_df['geometry'].area

# Convert 'Area_sqm' to a more suitable unit (e.g., square kilometers)
pop_df['Area_sqkm'] = pop_df['Area_sqm'] / 1e6  # 1 square kilometer = 1,000,000 square meters

# Calculate population density (people per square kilometer)
pop_df['Population Density (people/sqkm)'] = pop_df['TOTAL POPULATION'] / pop_df['Area_sqkm']
```

## Import Crime Counts & Other Datasets

```python
# SQL query or table name
query = "SELECT * FROM Chicago_Violent_Crime_Totals"  #Name of table 

# Create a DataFrame from the SQL query
violent_df = pd.read_sql(query, engine)
```

```python
housing_df = pd.read_csv('Affordable_Rental_Housing_Developments_20231109.csv')
grocery_stores = pd.read_csv('Grocery_Stores_-_2013_20231109.csv')
census_df = pd.read_csv('Census_Data_-_Selected_socioeconomic_indicators_in_Chicago__2008___2012_20231109.csv')
```

```python
#Census
census_df = census_df.rename(columns={'Community Area Number':'Community Area'})
census_df = census_df.drop(columns='COMMUNITY AREA NAME')

#Grocery
grocery_store_count = grocery_stores.groupby(['COMMUNITY AREA']).agg(Grocery_Store_Count=('STORE NAME','count')).reset_index()
grocery_store_count = grocery_store_count.rename(columns={'COMMUNITY AREA':'Community Area'})

#Housing
exclude_types = [
    'Artist Housing', 'Multifamily/Artists', 'Artist Live/Work Space', 
    'Artists & Families', 'Artist/Family', 'Senior HUD 202',
    'Senior LGBTQ',  
    'Inter-generational'
]

housing_df = housing_df[~housing_df['Property Type'].isin(exclude_types)]

housing_df.rename(columns={'Community Area Number':'Community Area'}, inplace=True)

columns_to_drop = ['Zip Code', 
                   'Phone Number', 
                   'Management Company', 
                   #'Community Area',
                   'Property Type',
                   'Property Name',
                   'Address',
                   'X Coordinate',
                   'Y Coordinate',
                   #'Latitude',
                   #'Longitude',
                   #'Community Area Name',
                   #'Location'
                  ]

housing_clean = housing_df.drop(columns=columns_to_drop)
housing_clean

affordable_housing_unit_count = housing_clean.groupby('Community Area')['Units'].sum().reset_index()
```

```python
merged_df = pd.merge(violent_df, affordable_housing_unit_count,on='Community Area',how='left')
merged_df = pd.merge(merged_df, grocery_store_count, on='Community Area',how='left')
merged_df = pd.merge(merged_df, census_df, on='Community Area',how='left')
```

```python
merged_df
```

```python
# Merge violent_df with population_df based on a common identifier, e.g., 'SECTOR' or 'CENSUS BLOCK FULL'
merged_df_pop = merged_df.merge(pop_df, left_on='Community Area', right_on='AREA_NUM_1', how='left')

# Calculate Violent Crimes per Capita per 1,000 people
merged_df_pop['Violent Crimes per Capita'] = (merged_df_pop['Violent Crimes'] / merged_df_pop['TOTAL POPULATION']) * 1000
```

```python
merged_df_pop = merged_df_pop.drop(columns='AREA_NUM_1')
```
